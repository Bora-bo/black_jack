```python
import random
from IPython.display import clear_output
from blackjack_art import *
#This function will pick a random card from the list
def deal_card():
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
    deal = random.choice(cards)
    return deal

#This function will sum the list of cards passed into its parameter
def calculate_score(cards):
    if sum(cards) == 21 and len(cards) == 2:
        return 0 #0 means blackjack
    
    if sum(cards) > 21 and 11 in cards:
        cards.remove(11)
        cards.append(1)
    return sum(cards)


def compare(user_total, dealer_total):
    #If the computer and user both have the same score, then it's a draw.
    if user_total == dealer_total:
        return "It is a draw!!"
        
    #if the computer has a blackjack (0), then the user loses.
    elif dealer_total == 0:
        return "Blackjack!!\nDealer wins"
    
    #If the user has a blackjack (0), then the user wins
    elif user_total == 0:
        return "Blackjack!!\nUser wins"
    
    #If the user_score is over 21, then the user loses.
    elif user_total > 21:
        return "BUST!! You lose."
    
    #If the computer_score is over 21, then the computer loses.
    elif dealer_total > 21:
        return "BUST!! You Win."
    
    #If none of the above, then the player with the highest score wins.
    elif dealer_total > user_total:
        return "Dealer wins."
        
    #if user_total > dealer_total:
    else:
        return "User wins."
        
        
#main code
def play_game():
    #This block will deal 2 cards for dealer and user
    user_cards = []
    dealer_cards = []

    for _ in range(2):
        user_cards.append(deal_card())
        dealer_cards.append(deal_card())

    print(logo)
    game_over = 'False'
    while game_over == 'False':
        #This block of code will pass the user and dealer's cards to the function calculat_score() and stop the game is the if user_cards or dealer_cards is a blackjack or above 21
        user_score = calculate_score(user_cards)
        dealer_score = calculate_score(dealer_cards)
        print(f"Your hand is: {user_cards}. Total is: {user_score}")
        print(f"Dealer's hand is: {dealer_cards[0]}.")

        if user_score == 0 or dealer_score == 0 or user_score > 21:
            game_over = 'True'

        #This block of code asks the user if they want to draw another card. If not it ends the game
        else:
            user_choise = input("Would you like to draw another card? Press 'y' for yes and 'n' for no: ")
            if user_choise == 'y':
                user_cards.append(deal_card())

            elif user_choise == 'n':
                print(f"Dealer's hand is: {dealer_cards}. Total = {dealer_score}")
                game_over = 'True'

    #This block of code will add a card to the dealer's hand if the dealer's total is less than 17 and stop adding to dealer's hand if dealer's total
    #is above 17 or 21
    while game_over == 'True' and dealer_score < 17:
        dealer_cards.append(deal_card())
        dealer_score = calculate_score(dealer_cards)

        if dealer_score > 21:
            #print(f"Delaer's hand is over 21: Hand {dealer_cards}. Total {dealer_score}")
            game_over = 'False'

    else:
        compare(user_score, dealer_score)
        print(f"User card and score: {user_cards, user_score}. Dealer cards and score: {dealer_cards, dealer_score}. {compare(user_score, dealer_score)}")
        game_over = 'False'

while input("Do you want to play a game. 'y' for Yes and 'n' for No: ") == 'y':
    clear_output(wait=True)
    play_game()
            

    #Hint 14: Ask the user if they want to restart the game. If they answer yes, clear the console and 
    #start a new game of blackjack and show the logo from art.py.

```
